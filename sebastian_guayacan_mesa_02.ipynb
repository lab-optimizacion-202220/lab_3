# PRIMER PUNTO


```python
def capicula (num):
    ans = 0
    rev = str (num)[::-1]
    if (rev == str(num)):
        ans = 1
    return ans
```

```python
capicula(212)
```

```python
capicula(2332)
```

```python
capicula(12345678987654321)
```

```python
capicula (555555)
```

```python
import time
def capiculas (num):
    ans = 0
    aux = 0
    start = time.time()
    for i in range(num+1) :
        if capicula(aux) == 1 and aux >10:
            ans = ans + 1
        
        aux  = aux + 1
    end = time.time()
    print (f"Tiempo: {end - start}")
    return ans
```

```python
capiculas (10000000) 
```

# SEGUNDO PUNTO


```python
def ordenar_listas (A,B):
    ans = A + B
    ans.sort()
    return ans
    
```

```python
ordenar_listas([1, 3 , 5,7,9 ], [2,4,6.5,8,10])
```

# CUARTO PUNTO

```python

import sympy as sym
sym.init_printing()
a,b,c = sym.symbols('a,b,c')
#sistema de acuaciones planteado:
f = sym.Eq(2*a + 3*b + 4*c, 80)
g = sym.Eq(2*a + 2*b + 3*c, 60)

sol = sym.solve([f,g],(a,b,c))
print (f"Soluci√≥n al sistema {sol} ")
```

# QUINTO PUNTO

```python
#sistema de ecuaciones planteado para el spline cuadr√°tico:
# y = a + b*x + c*x¬≤
sym.init_printing()
a,b,c = sym.symbols('a,b,c')
f = sym.Eq(a + b + c, 3)
g = sym.Eq(a + 2*b + 4*c, 4)
h = sym.Eq(a + 3*b + 6*c, 7)

sol = sym.solve([f,g,h],(a,b,c))
print (f"Soluci√≥n al sistema {sol} ")
```
